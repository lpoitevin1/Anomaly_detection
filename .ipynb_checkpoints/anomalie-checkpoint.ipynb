```python
import numpy as np 
np.set_printoptions(threshold=10000,suppress=True) 
import pandas as pd
import warnings
import matplotlib.pyplot as plt 
warnings.filterwarnings('ignore')
```

# Mouse

```python
Mouse=pd.read_csv('./mouse.txt',sep=' ',header=None)
```

```python
Mouse.head(3)
```

```python
X_Mouse=Mouse.values
```

```python
plt.plot(X_Mouse[:,0],X_Mouse[:,1],'ro')
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
```

```python
IF=IsolationForest(n_estimators=500,contamination=0.01)
IF.fit(X_Mouse)
anomalie=IF.predict(X_Mouse)
```

```python
IF=IsolationForest(n_estimators=500)
IF.fit(X_Mouse)
decision=IF.decision_function(X_Mouse)
decision

```

```python
plt.figure(figsize=(8, 6))
plt.hist(decision, bins=50)
plt.show()
```

```python
anomalie=decision.copy()
anomalie[decision<=-0.08]=1
anomalie[decision>-0.08]=0
```

```python
plt.scatter(X_Mouse[:,0],X_Mouse[:,1],c=anomalie, cmap=plt.get_cmap('spring'))
plt.show()
```
